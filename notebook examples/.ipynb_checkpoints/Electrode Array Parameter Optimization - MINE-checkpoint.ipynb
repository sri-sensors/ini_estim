# Electrode Array Parameter Optimization with MINE
In this notebook, we demonstrate the optimization of bias and gain parameters for the output of an electrode array given input stimulus ginals that follow a Gaussian distribution. To perform the optimization, we maximize mutual information following the techniques of [1] and [2].

1. Hjelm, R. Devon, et al. "Learning deep representations by mutual information estimation and maximization." arXiv preprint arXiv:1808.06670 (2018).
2. Belghazi, Mohamed Ishmael, et al. "Mutual information neural estimation." International Conference on Machine Learning. 2018.

```python
# Import all of the necessary modules
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
import npeet.entropy_estimators as ee
from ini_estim.electrodes import electrode_arrays
from ini_estim.ml.electrode_array import ElectrodeArray
from ini_estim.ml.nerve import Nerve
from ini_estim.ml.encoder import train_mi_encoder
from ini_estim.ml.loss import InfoMaxLoss
from ini_estim.ml.discriminator_models import MLPDiscriminator
from ini_estim.mutual_information.nn import mi_ksg, normalize_data
```

## Define the forward model
In PyTorch, we must define a module that takes the input (Gaussian data) and computes the output (neuron activation probabilities). We can then optimize the module parameters (gain and bias on the input) using stochastic gradient descent or one of its variations.

```python
# PyTorch module that we are optimizing
#   This module applies a gain and bias to the input
#   before applying to the electrode array. Final
#   outputs are the activation probabilities given by
#   the array.
class Model(nn.Module):
    def __init__(self, electrode_array):
        super().__init__()
        self.gain = nn.Parameter(torch.rand(1)*0.05, requires_grad=True)
        self.bias = nn.Parameter(torch.rand(1)*0.05, requires_grad=True)
        self.ea = electrode_array
        self.in_features = electrode_array.in_features
        self.out_features = electrode_array.out_features
    
    def forward(self, x):
        # 'encoder' - gain and bias change the effective variance/mean
        x = self.gain*x + self.bias
        # final output is the electrode array activation probabilities
        return self.ea(x)
```

## Define the electrode array and optimization parameters
For this example, we have chosen a generic 7-electrode round nerve cuff. The ElectrodeArray class is a parameterized PyTorch module that will take the defined nerve cuff and produce neuron activation probalities, for 500 uniformly distributed neurons by default. We also define the loss function, which takes as its input an auxiliary neural network for its estimation of Mutual Information, or in this case Jensen-Shannon Divergence. Finally, the training data is defined by sampling from the Normal/Gaussian distribution.

```python
# Define our electrode array
ea = electrode_arrays['generic_cuff']()
ea.active_electrodes = ea.electrodes[::2]

# Instantiate our end-to-end model to optimize.
model = Model(ElectrodeArray(ea))
# Define the loss function for the optimizer
loss = InfoMaxLoss(MLPDiscriminator(model.in_features, model.out_features))
# Define our dataset
batch_size = 64
nbatches = 10
data = torch.randn(nbatches*batch_size, model.in_features)
data_loader = DataLoader(data, batch_size)
```

## Optimization
In this step, we optimize according to the loss function defined above. Independently, we compute the mutual information using [3] to verify that it is being maximized. 

3. Kraskov, Alexander, Harald St√∂gbauer, and Peter Grassberger. "Estimating mutual information." Physical review E 69.6 (2004): 066138.

```python
# Perform the optimization
num_epochs=250
gain_history = []
bias_history = []
mi_history = []

# Define a callback to independently compute the Mutual Information at each step.
def callback(epoch, mymodel):
    # sub-sample the data and compute the response
    data_sample = data[:5*batch_size]
    with torch.no_grad():
        nerve_out = mymodel(data_sample)
        
    gain_history.append(mymodel.gain.item())
    bias_history.append(mymodel.bias.item())
    # Scale the activation probabilities. In theory, this should 
    # have no effect on the calculated mutual information. However,
    # the KSG estimator performs poorly when the input and output 
    # have different value ranges. Since the probability of 
    # activation is in range [0, 1], we scale it so that if it 
    # followed a uniform distribution, its standard deviation would be
    # approximately the same as the input gaussian distribution.
    yt = 3.464*(nerve_out.numpy().astype(np.float)-0.5)
    xt = data_sample.numpy().astype(np.float)
    miest = mi_ksg(xt, yt, normalize=False)
    mi_history.append(miest)

# get initial MI
callback(-1, model)
# optimize
loss_history, _, _ = train_mi_encoder(model, loss, data_loader, num_epochs=num_epochs, callback=callback, subpbar=False)
```

## Plot the optimization history

```python
# Show the optimization results
plt.figure()
plt.plot(loss_history)
plt.grid()
plt.xlabel("Epoch")
plt.ylabel("Jensen-Shannon Divergence Loss")
plt.title("Training Loss vs. Epoch")

plt.figure()
plt.plot(np.array(mi_history)/np.log(2))
plt.xlabel("Epoch")
plt.ylabel("Mutual Information (bits)")
plt.title("Mutual Information vs. Epoch")
plt.grid()

plt.figure()
plt.plot(gain_history, label="Standard Deviation")
plt.plot(bias_history, label="Mean")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("Parameters vs. Epoch")
plt.grid()
plt.legend()
plt.show()
```

## Animation of the electrode cuff activation probabilities

```python
from matplotlib import animation, rc
from matplotlib import cm
from IPython.display import HTML
cmap = cm.get_cmap()
fig, ax = plt.subplots(1,2, figsize=[8,4])

with torch.no_grad():
    data_out = model(data)
    
scatter = ax[0].scatter(model.ea.x, model.ea.y)
ax[0].set_title("Electrode Cuff Activation Probabilities")
ax[0].set_xlabel("X location (mm)")
ax[0].set_ylabel("Y location (mm)")
ax[0].axis("equal")
xstem = np.arange(model.ea.in_features)
h_stem = ax[1].stem(xstem, np.zeros(model.ea.in_features), use_line_collection=True)
a = model.gain.item()
b = model.bias.item()
ax[1].set_ylim([b - 4*(a), b + 4*(a)])
ax[1].grid()
ax[1].set_xlabel("Electrode #")
ax[1].set_ylabel("Input Current (mA)")
ax[1].set_title("Electrode Inputs")
plt.subplots_adjust(hspace=0.7, wspace=0.4)
def animate(i):
    scatter.set_facecolor(cmap(data_out[i,:].numpy()))
    y = data[i].numpy()*model.gain.item() + model.bias.item()
    h_stem[0].set_ydata(y)

    # stemlines
    h_stem[1].set_paths(
        [np.array([[xstem[j], 0], [xstem[j], y[j]]]) for j in np.arange(model.ea.in_features)])

anim = animation.FuncAnimation(fig, animate, frames=batch_size, interval=100)
HTML(anim.to_html5_video())
```

## Show the average activation probabilities

```python
dmean = data.mean(dim=0, keepdim=True)
with torch.no_grad():
    pact = model(dmean).squeeze()
plt.figure()
plt.scatter(model.ea.x, model.ea.y, s=150, c=pact, vmin=0, vmax=1)
plt.axis("equal")
plt.ylabel("Y Position (mm)")
plt.xlabel("X Position (mm)")
plt.title("Probability of activation at mean signal level")
plt.colorbar()
plt.show()
```
