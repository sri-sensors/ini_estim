# Electrode Array Parameter Optimization
In this notebook, we demonstrate the optimization of bias and gain parameters for the output of an electrode array given input stimulus signals that follow a Gaussian distribution

```python
# Import all of the necessary modules
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import tqdm.notebook as tqdm
import ini_estim.electrodes as electrodes
import ini_estim.nerve_models as nerve_models
import ini_estim.sensors as sensors
import ini_estim.mutual_information.nn as mi
import ini_estim.utilities.data as datautils
from ini_estim.utilities.sampling import uniform_points_circle
```

## Define the electrode array and nerve geometry 
For this example, we have chosen a generic 7-electrode round nerve cuff with a 3mm diameter. The neuron responses follow a probabilistic model, where the neuron diameters are assumed to have a Gaussian distribution. Following this distribution, the nerve model derives activation probabilities for given stimulus potential values. 

```python
# Define our electrode array and nerve response model
electrode_array = electrodes.GenericElectrodeCuff()
nerve_model = nerve_models.Nerve()

# The default GenericElectrodeCuff has 14 electrodes, but this tends to be too 
# many for this kind of optimization, so we will simulate a 7-electrode cuff 
# by taking every other electrode.
num_electrodes = 7

# We must also define the number of neurons to simulate. For this example, 
# we choose 500. While a real nerve could have thousands, this would take 
# a long time to optimize, and in practice, doesn't make much of a difference 
# for the final answer. 
num_neurons = 500

# The electrode array class defines a full potential map for each electrode.
# For display purposes, we'll take the average potential map
electrode = list(range(0, electrode_array.num_electrodes, electrode_array.num_electrodes//num_electrodes))
electrode_labels = ["Electrode {}".format(i+1) for i in range(num_electrodes)]
electrode_map, xe, ye = electrode_array.get_potential_map(electrode, ds=0.01)
electrode_map = np.mean(electrode_map, axis=0)

# Place the neurons uniformly around the electrod region
xn, yn = uniform_points_circle(num_neurons, electrode_array.DIAMETER_MM)
# Get the weights for converting stimulus current values to potentials for each neuron.
electrode_weights = np.column_stack(electrode_array.get_potential(xn, yn)).transpose()
electrode_weights = electrode_weights[electrode, :]

plt.figure()
plt.imshow(electrode_map, extent=electrode_array.extent)
plt.scatter(xn, yn, c="orange", label="neuron positions")
plt.xlim(electrode_array.extent[0:2])
plt.ylim(electrode_array.extent[2:4])
plt.xlabel("Position (mm)")
plt.ylabel("Position (mm)")
plt.title("Electrode Array and Neuron Positions")
plt.legend()

plt.show()
```

## Define the signal parameters
Each electrode has an input signal following a zero-mean Gaussian distribution with a standard deviation of 1. We also define a peak signal-to-noise ratio, which coarsely models the probability that each neuron has a non-zero probability of random (unstimulated) activation. 

```python
# Define an input signal for each electrode and peak signal-to-noise ratio
num_samples = 1000
psnr_db = 20

psnr = 10**(psnr_db/20)
usig = np.random.randn(num_samples, len(electrode))
unoise = np.random.randn(num_samples, num_neurons)/psnr

plt.figure()
plt.hist(usig)
plt.grid()
plt.xlabel("Value")
plt.ylabel("Count")
plt.legend(electrode_labels)
plt.title("Input Signal Histogram")
plt.show()
```

## Define the optimization loss function
We optimize the electrode array parameters by maximizing the mutual information between the output signal and input signal. To measure mutual information for any particular set of parameters, we use the KSG[1] non-parametric estimator. 

[1] A. Kraskov, H. Stogbauer and P. Grassberger, "Estimating mutual information". Phys. Rev. E 69, 2004.

```python
# Define the optimization loss function using mutual information
def mi_loss(params, nerve_model, usig, noise, history=None):
    gain = params[0]
    bias = params[1]
    y_in = gain*usig + bias
    
    # convert the electrode stimulus current to local potential 
    # and add the noise
    e_in = y_in @ electrode_weights
    p_act = nerve_model.get_activation_probability(e_in)
    p_act += noise
    p_act = np.clip(p_act, 0, 1)
    
    # Scale the activation probabilities. In theory, this should 
    # have no effect on the calculated mutual information. However,
    # the KSG estimator performs poorly when the input and output 
    # have different value ranges. Since the probability of 
    # activation is in range [0, 1], we scale it so that if it 
    # followed a uniform distribution, its standard deviation would be
    # approximately the same as the input gaussian distribution.
    p_act -= 0.5
    p_act *= 3.464
    
    # Scaling is for standard deviation = 1, so scale the signal too.
    minfo = mi.mi_ksg(usig/np.std(usig, axis=0), p_act)
    if history is not None:
        history.append(minfo)
        
    # we want to maximize mutual information, but optimizers
    # do minimization, so return negative value.
    return -minfo

# Coarse grid search to find good starting point
gain = np.linspace(0.01, 0.1, 10)
bias = np.linspace(0.0, 0.1, 10)
A,B = np.meshgrid(gain, bias)
MI = np.zeros((gain.size, bias.size))
for i, ab in tqdm.tqdm(list(enumerate(zip(A.ravel(), B.ravel())))):
    r = i % gain.size
    c = i // bias.size
    MI[r, c] = -mi_loss(ab, nerve_model, usig, unoise)


imax = np.argmax(MI)
rmax, cmax = np.unravel_index(imax, MI.shape)
gain_max = gain[rmax]
bias_max = bias[cmax]
print("Peak mutual information: {} bits".format(MI.ravel()[imax]))
print("Gain: {}, Bias: {}".format(gain_max, bias_max))

plt.figure()
plt.imshow(MI, extent=[bias[0], bias[-1], gain[-1], gain[0]], aspect='auto')
plt.colorbar()
plt.xlabel("Bias")
plt.ylabel("Gain")
plt.title("Mutual Information")
plt.show()
```

## Final optimization
In the previous step, we performed a coarse grid search to find a starting point for the optimization. To perform the final optimization, we use scipy's minimize function with the Nelder-Mead simplex method. This method was chosen because the KSG estimator doesn't provide gradient information, so we are limited to gradient-free optimization algorithms. 

```python
param_history = []
loss_history = []
def cback(xk, state=None):
    param_history.append(xk)
    
res = opt.minimize(mi_loss, [gain_max, bias_max], (nerve_model, usig, unoise, loss_history), 
                   method="Nelder-Mead", callback=cback, 
                   options={'disp': True, 'adaptive': True})


gain_f, bias_f = res['final_simplex'][0][0]
print("Optimized gain: {}\nOptimized offset: {}".format(gain_f, bias_f))

plt.figure()
plt.plot(loss_history)
plt.xlabel("Evaluation Iteration")
plt.ylabel("Mutual Information")
plt.grid()
plt.title("Mutual Information vs. Evaluation Iteration")

plt.figure()
plt.plot(np.array(param_history))
plt.grid()
plt.legend(["Gain", "Offset"])
plt.xlabel("Iteration")
plt.ylabel("Parameter Value")
plt.title("Parameters vs. Iteration")

y_in = gain_f*usig + bias_f
plt.figure()
plt.hist(y_in)
plt.grid()
plt.xlabel("Value")
plt.ylabel("Count")
plt.legend(electrode_labels,
          bbox_to_anchor=(1.0, 1.0))
plt.title("Electrode stimulus with final parameters")

plt.show()
```

## Plot the activation probabilities using the final parameters

```python
def get_activation_probability(signal_level, e=0):
    sig = gain_f*signal_level + bias_f
    e_in = sig*electrode_weights
    e_all = np.sum(e_in, axis=0)
    p_all = nerve_model.get_activation_probability(e_all)
    p_single = nerve_model.get_activation_probability(e_in[e,:])
    return p_all, p_single


p_max = get_activation_probability(np.max(usig))
p_avg = get_activation_probability(np.mean(usig))
p_min = get_activation_probability(np.min(usig))

labels = ["max", "mean", "min"]
cnorm_lin = plt.Normalize(0, 1, True)
cnorm_log = colors.LogNorm(1e-3, 1, True)

for p, l in zip([p_max, p_avg,p_min], labels):
    plt.figure(figsize=[12,4])
    plt.subplot(121)
    plt.scatter(xn, yn, s=150, c=p[0], norm=cnorm_log)
    plt.xlim([-1, 1])
    plt.ylim([-1, 1])
    plt.axis("equal")
    plt.colorbar()
    plt.title("Probability of activation, all electrodes at {} signal".format(l))
    
    plt.subplot(122)
    plt.scatter(xn, yn, s=150, c=p[1], norm=cnorm_log)
    plt.xlim([-1, 1])
    plt.ylim([-1, 1])
    plt.axis("equal")
    plt.colorbar()
    plt.title("Probability of activation, single electrode at {} signal".format(l))


plt.show()
```
